```python
from pycaret.datasets import get_data
data = get_data('boston')
```

```python
corr = data.corr()
corr
```

```python
from pycaret.regression import *
s = setup(data, target = 'medv', session_id = 123)
```

```python
compare_models()
```

```python
best = compare_models(fold=15)
```

```python
plot_model(best)
```

```python
evaluate_model(best)
```

```python
predictions = predict_model(best, data=data)
```
